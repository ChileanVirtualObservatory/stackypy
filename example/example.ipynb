```python
import stackypy
import pyfits
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sp0 = pyfits.open('spiral/000.fits')[0].data
(c_x,c_y,angle,ma_axis,mi_axis) = stackypy.detect_object(sp0)
print("center:    (%f,%f)"%(c_x,c_y))
print("angle:      %f"%angle)
print("major axis: %f"%ma_axis)
print("minor axis: %f"%mi_axis)

plt.imshow(sp0,cmap='gray')
plt.show()
# 
detec = stackypy.get_detection(sp0)
plt.imshow(detec,cmap='gray')
plt.show()
```

```python
imgs = []
for i in range(18):
    imgs.append(pyfits.open('spiral/%03d.fits'%i)[0].data)
temp,used = stackypy.stack_to_template(imgs)
print(used)
plt.imshow(temp,cmap='gray')
plt.show()
```
